## Step 3: Preprosessing

I need to one-hot encode the classifier variables as well as scale everything so that we get some nice ol' compareable features!

```python
# Time to get some heavy duty hitters in here

from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split, cross_validate, GridSearchCV, learning_curve
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
```

```python
drop.columns
```

```python
drop1 = drop.drop(columns=['PICKUP_DATE','ORDER_BOOK_DATE'])
drop1.head()
```

```python
# Create a new dataframe for the one-hot encoded variables

fin = pd.get_dummies(drop1, columns=['HAZMAT',
       'SERVICE_LEVEL', 'EQUIP_TYPE',
       'ORIGIN_LOCATION', 'DESTINATION_LOCATION', 'date'], prefix=['H', 'SL', 'ET', 'OL', 'DL', 'D'])
```

```python
# Check one more time for null values
fin.isnull().sum()
```

```python
#Make feature and target arrays

X = fin.drop(columns='SHIPMENT_AMOUNT', axis=1)
y = fin['SHIPMENT_AMOUNT']
Xcol = X.columns
```

```python
#Scale the features

scaler = StandardScaler()
X = scaler.fit_transform(X)
```

```python
# Split the fin dataset into train and tests so that we can evaluate the model

X_train, X_test, y_train, y_test = train_test_split(X, 
                                                    y, test_size=0.3, 
                                                    random_state=47)
```

```python
# Let's try a model

from sklearn.ensemble import RandomForestRegressor
rfr = RandomForestRegressor(n_estimators=300, random_state = 1,n_jobs=-1)
model_res = rfr.fit(X_train, y_train)
y_pred = model_res.predict(X_test)
```

```python
# Evaluate the model

mse = mean_squared_error(y_test, y_pred)
print(f'Mean Squared Error: {mse}')
 
r2 = r2_score(y_test, y_pred)
print(f'R-squared: {r2}')
```

```python
rfr.feature_importances_
```

```python
coefficients = rfr.feature_importances_
feature_importance = pd.DataFrame({'Feature': Xcol, 'Coefficient': coefficients})
feature_importance = feature_importance.reindex(feature_importance['Coefficient'].abs().sort_values(ascending=True).index)
print(feature_importance)
```

```python
plt.figure(figsize=(10, 20))
plt.barh(feature_importance['Feature'], feature_importance['Coefficient'], color='skyblue')
plt.xlabel('Coefficient Value')
plt.title('Feature Importances in Random Forest Regression')
plt.grid(axis='x', linestyle='--', alpha=0.6)
plt.show()
```
