```python
import xgboost as xgb
from sklearn.model_selection import train_test_split
```

```python
binary_question = question
binary_question['income'].replace({'<=50K':0, '>50K':1}, inplace=True)
```

```python
binary_question.drop(columns=['workclass', 'occupation'], inplace=True)
binary_question
```

```python
binary_question['native-country'].value_counts()
```

```python
binary_question = pd.get_dummies(binary_question)
```

```python
X, y = binary_question.iloc[:, :-1], binary_question.iloc[:, -1]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, 
                                                    test_size=0.2, random_state=24)
```

```python
#Instantiate with hyperparameters 
xg_cl = xgb.XGBClassifier(objective='binary:logistic', 
                          n_estimators=10, seed=24)
```

```python
xg_cl.fit(X_train, y_train)
```

```python
preds=xg_cl.predict(X_test)
```

```python
accuracy = float(np.sum(preds==y_test))/y_test.shape[0]
print('accuracy: %f' %(accuracy))
```

```python
booster = xg_cl.get_booster()
importance_dict = booster.get_score(importance_type='weight')
```

```python
for feature, importance in importance_dict.items():
    print(f'{feature}: {importance}')
```

```python
xgb.plot_importance(xg_cl, importance_type='gain')
plt.show()
```

```python
#FoR Cross Validation, you need a DMatrix

binary_dmatrix=xgb.DMatrix(data=binary_question.iloc[:, :-1], label=binary_question.iloc[:, -1])

#Also need the parameters defined

params={'objective':'binary:logistic', 'max_depth':4}
```

```python
#nfold is the number of cross_validation folds we're going to do
#num_boost_rounds is the number of trees

cv_results = xgb.cv(dtrain=binary_dmatrix, params=params, nfold=4,
                   num_boost_round=10, metrics='error', as_pandas=True)
```

```python
print('Accuracy: %f' %((1-cv_results['test-error-mean']).iloc[-1]))
```

```python
#Find the AUC

cv_results1 = xgb.cv(dtrain=binary_dmatrix, params=params, nfold=4,
                   num_boost_round=5, metrics='auc', as_pandas=True,
                    seed=2)

print(cv_results1)
print((cv_results1['test-auc-mean']).iloc[-1])
```

Loss functions in XGBoost

Regression: reg:linear 

Classification (decisions, not prob): reg:logistic 

Classification (probabilty of the positive class): binary:logistic
