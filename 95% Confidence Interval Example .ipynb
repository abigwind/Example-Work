```python
import scipy.stats
import scipy.optimize
import scipy.spatial
```

```python
pop = pd.read_csv('grades.csv')
```

Create a histogram of the grade distribution 

```python
pop.grade.hist(histtype='step')
```

We want to find out what percentile gave a 5 grade.

```python
scipy.stats.percentileofscore(pop.grade, 5)
```

Here is the bootstrap distribution using sampling with replacement 1000 times.

```python
bootstrap = pd.DataFrame({'passing': [scipy.stats.percentileofscore(pop.sample(100, replace=True).grade,5) for i in range(1000)]})
```

Plot the bootstrap histogram

```python
bootstrap.passing.hist(histtype='step')
```

Then find the 95% interval between 2.5% and 97.5% percentiles

```python
bootstrap.passing.quantile(0.025), bootstrap.passing.quantile(0.975)
```

The 95% Confidence Interval for the population to grade a 5 is the 33 to 53 percentile
