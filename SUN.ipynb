```python
# List of days
days = ['Monday', 'Tuesday', 'Wednesday', 'Thursday']

# Initialize lists to store counts of 1st and last PM activities for each day
counts_1st_pm = []
counts_last_pm = []

# Iterate over each day and calculate counts
for day in days:
    # Extract columns related to the current day
    day_columns = [col for col in after.columns if day in col]
    # Extract 1st and last PM activities for the current day
    day_1st_pm = after[f'{day} - 1st PM Activity'].notnull().sum()
    day_last_pm = after[f'{day} - Last PM Activity'].notnull().sum()
    # Append counts to the lists
    counts_1st_pm.append(day_1st_pm)
    counts_last_pm.append(day_last_pm)

# Create a DataFrame with counts for each day
counts_df = pd.DataFrame({
    'First Activity': counts_1st_pm,
    'Last Activity': counts_last_pm
}, index=days)

# Plot stacked bar plot
ax = counts_df.plot(kind='bar', stacked=True, color=['skyblue', 'lightgreen'], figsize=(10, 6))
plt.title('Count of Students for First and Last PM Activities by Day')
plt.xlabel('Day')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Activity')
plt.tight_layout()
plt.show()
```

```python
unique_class_counts = {}

# Iterate over each day and count unique classes
for day in ['Monday - 1st PM Activity', 'Monday - Last PM Activity',
       'Tuesday - 1st PM Activity', 'Tuesday - Last PM Activity',
       'Wednesday - 1st PM Activity', 'Wednesday - Last PM Activity',
       'Thursday - 1st PM Activity', 'Thursday - Last PM Activity',
       'Friday - 1st PM Activity', 'Friday - Last PM Activity']:
    # Concatenate data for the day and remove missing values
    day_data = after[day].dropna()
    # Count unique classes
    unique_class_counts[day] = len(day_data.unique())

# Create a bar chart for unique class counts for each day
plt.figure(figsize=(10, 6))
plt.bar(unique_class_counts.keys(), unique_class_counts.values(), color='skyblue')
plt.title('Unique Classes Taken Each Day')
plt.xlabel('Day')
plt.ylabel('Number of Unique Classes')
plt.xticks( rotation=45, ha='right')

#plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```

```python
class_counts = {}

# Iterate over each day and count occurrences of each class
for day in ['Monday - 1st PM Activity', 'Monday - Last PM Activity',
       'Tuesday - 1st PM Activity', 'Tuesday - Last PM Activity',
       'Wednesday - 1st PM Activity', 'Wednesday - Last PM Activity',
       'Thursday - 1st PM Activity', 'Thursday - Last PM Activity',
       'Friday - 1st PM Activity', 'Friday - Last PM Activity']:
    # Concatenate data for the day and remove missing values
    day_data = after[day].dropna()
    # Count occurrences of each class
    for class_name in day_data:
        if class_name not in class_counts:
            class_counts[class_name] = 1
        else:
            class_counts[class_name] += 1

# Print the counts for each unique class
for class_name, count in class_counts.items():
    print(f"{class_name}: {count} students")
```

```python
plt.figure(figsize=(10, 6))
plt.bar(class_counts.keys(), class_counts.values(), color='skyblue')
plt.title('Number of Students in Each Class')
plt.xlabel('Class')
plt.ylabel('Number of Students')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
class_counts_sorted = dict(sorted(class_counts.items(), key=lambda x: x[1], reverse=True))
average_kids_per_class = sum(class_counts.values()) / len(class_counts)

# Create a bar graph for each class and their counts in descending order
plt.figure(figsize=(10, 6))
plt.bar(range(len(class_counts_sorted)), class_counts_sorted.values(), color='skyblue')
plt.title('Number of Students in Each Class (Descending Order)')
plt.xlabel('Class')
plt.ylabel('Number of Students')

# Set x-axis tick positions and labels
plt.xticks(range(len(class_counts_sorted)), class_counts_sorted.keys(), rotation=45, ha='right')
plt.axhline(y=average_kids_per_class, color='red', linestyle='--', label=f'Average ({average_kids_per_class:.2f} students)')

plt.grid(axis='y')
plt.tight_layout()
plt.show()
```
